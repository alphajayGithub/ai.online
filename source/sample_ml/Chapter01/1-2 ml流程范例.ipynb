# 有监督学习（分类）的例子

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
data = load_breast_cancer()

cancers = load_breast_cancer()
X = cancers.data       #获取特征值
Y = cancers.target     #获取标签
print(X.shape)         #查看特征形状
print(Y.shape)         #查看标签形状

#print(X)
#print(Y)

#dict_keys(['data', 'target', 'target_names', 'DESCR', 'feature_names', 'filename'])
print(cancers.keys())
print(data.DESCR)  #查看数据集描述
```

```python
print('特征名称')
print(data.feature_names)  # 特征名
print('分类名称')
print(data.target_names)  # 标签类别名

# 注意返回值： 训练集train，x_train，y_train，测试集test，x_test，y_test
# x_train为训练集的特征值，y_train为训练集的目标值，x_test为测试集的特征值，y_test为测试集的目标值
# 注意，接收参数的顺序固定
# 训练集占80%，测试集占20%
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print('训练集的特征值和目标值：', x_train, y_train)
print('测试集的特征值和目标值：', x_test, y_test)
```

```python
X = data.data
y = data.target
X_pd = pd.DataFrame(data.data, columns=data.feature_names)
X_pd.head()
print(X_pd.head(1))

print("before slice", X.shape)
X = X[:, :10]
print("after slice", X.shape)

```

## 实现方法

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000)  
#  https://www.cnpython.com/qa/1283852  set max_iter to avoid lbfgs problem
```

```python
# x_train, x_test, y_train, y_test
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print(y_test)

print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

# 无监督学习（聚类）的例子

```python
from sklearn.datasets import load_wine
data = load_wine()

print(data.keys())



print(data.DESCR)
print(data.feature_names)
print(data.target_names)

X_pd = pd.DataFrame(data.data, columns=data.feature_names)
X_pd.head()
```

```python
X = data.data[:, [0, 9]]  # alcohol + color_intensity , use as x/y axis for 2-D diagram
print(X.shape)
```

## 实现方法

```python
from sklearn.cluster import KMeans
n_clusters = 4
kmodel = KMeans(n_clusters=n_clusters)
```

```python
pred = kmodel.fit_predict(X)
```

## 查看结果

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(X[pred==0, 0], X[pred==0, 1], color='red', marker='s', label='Label1')
ax.scatter(X[pred==1, 0], X[pred==1, 1], color='blue', marker='s', label='Label2')
ax.scatter(X[pred==2, 0], X[pred==2, 1], color='green', marker='s', label='Label3')
ax.scatter(X[pred==3, 0], X[pred==3, 1], color='pink', marker='s', label='Label4')  # add this for 4th class
ax.scatter(kmodel.cluster_centers_[:, 0], kmodel.cluster_centers_[:, 1], s=200, color='yellow', marker="*", label="center")
ax.legend()
plt.show()
```

# 图形的种类和画法：使用Matplotlib显示图形的方法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-5, 5, 101)
y1 = np.sin(x1)
print(x1)
print(y1)
```

```python
plt.plot(x1, y1)
```

```python
fig, ax = plt.subplots()
ax.set_title("Sin")
ax.set_xlabel("rad")
ax.plot(x1, y1)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.show()
```
